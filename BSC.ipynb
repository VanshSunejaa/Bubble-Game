```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import re

# Load existing CSV
df = pd.read_csv("BSC_PRODUCTS.csv")  # Replace with your file name

# Store actual ratings here
actual_ratings = []

# Loop through each product URL
for url in df['Product URL']:
    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
        soup = BeautifulSoup(response.text, "lxml")

        # Get all tags with the rating class
        rating_tags = soup.find_all("div", class_="css-m6n3ou")

        # Filter only text that looks like a rating (e.g., 4.5/5)
        rating = "No Rating"
        for tag in rating_tags:
            text = tag.get_text(strip=True)
            if re.match(r"^\d+(\.\d+)?\/5$", text):  # Matches formats like 4.3/5
                rating = text
                break

        actual_ratings.append(rating)
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        actual_ratings.append("Error")

    # Respectful scraping delay
    time.sleep(1)

# Add the cleaned actual ratings to the dataframe
df["Actual Rating"] = actual_ratings

# Save to a new CSV
df.to_csv("bsc_with_clean_ratings.csv", index=False)

print("✅ Actual ratings added successfully.")
```

```python
import pandas as pd

# Load the file
df = pd.read_csv('BSC_Amazon_Product_Dataset.csv')

# Preview the first few rows
df.head()
```

```python
import re

def convert_sales_volume(text):
    if isinstance(text, str):
        # Extract number using regex
        match = re.search(r'(\d+)([Kk]?)\+', text)
        if match:
            num = int(match.group(1))
            if match.group(2).lower() == 'k':
                return num * 1000
            return num
    return None

# Apply the function to the 'Sales Volume' column
df['Monthly Sales'] = df['Sales Volume'].apply(convert_sales_volume)
```

```python
# Calculate discount where original price is available
df['Discount %'] = ((df['Price Strikethrough (INR)'] - df['Price (INR)']) / df['Price Strikethrough (INR)']) * 100

# Round off to 1 decimal place
df['Discount %'] = df['Discount %'].round(1)
```

#Categorizing Data

```python
def extract_category(title):
    title = title.lower()
    if "trimmer" in title:
        return "Trimmer"
    elif "shaving" in title:
        return "Shaving"
    elif "face wash" in title or "scrub" in title or "skincare" in title or "mask" in title:
        return "Skincare"
    elif "perfume" in title or "body spray" in title or "fragrance" in title:
        return "Fragrance"
    elif "gift" in title or "kit" in title:
        return "Gift Kit"
    elif "soap" in title or "bath" in title:
        return "Bath & Body"
    elif "hair removal" in title:
        return "Hair Removal"
    else:
        return "Other"

df['Category'] = df['Title'].apply(extract_category)

```

```python
import pandas as pd
import numpy as np

# Load datasets
df_bsc = pd.read_csv("BSC_Amazon_Product_Dataset.csv")
df_beardo = pd.read_csv("beardo data.csv")

# Standardize column names
df_bsc = df_bsc.rename(columns={
    "Price Strikethrough (INR)": "Original Price (INR)",
    "Reviews Count": "Review Count",
    "Sales Volume": "Monthly Sales"
})

df_bsc['Brand'] = 'BSC'
df_beardo['Brand'] = 'Beardo'

# Define common structure
common_cols = ['Title', 'Brand', 'Price (INR)', 'Original Price (INR)', 'Discount %',
               'Rating', 'Review Count', 'Monthly Sales']

# Ensure all columns exist
for col in common_cols:
    if col not in df_bsc.columns:
        df_bsc[col] = None
    if col not in df_beardo.columns:
        df_beardo[col] = None

df_bsc = df_bsc[common_cols]
df_beardo = df_beardo[common_cols]

# Combine datasets
df_trimmers = pd.concat([df_bsc, df_beardo], ignore_index=True)

# Clean numeric columns
for col in ['Price (INR)', 'Original Price (INR)', 'Discount %', 'Rating', 'Review Count', 'Monthly Sales']:
    df_trimmers[col] = pd.to_numeric(df_trimmers[col].replace('[^0-9.]', '', regex=True), errors='coerce')

# 🚨 Drop rows where Price is missing or zero
df_trimmers = df_trimmers[df_trimmers['Price (INR)'].notna() & (df_trimmers['Price (INR)'] > 0)]

# 🔄 Recreate Price Band
bins = [0, 1000, 1500, 2000, np.inf]
labels = ['Under ₹1K', '₹1K–1.5K', '₹1.5K–2K', '₹2K+']
df_trimmers['Price Band'] = pd.cut(df_trimmers['Price (INR)'], bins=bins, labels=labels)

# 🧾 Group by Brand and Price Band
price_band_summary = df_trimmers.groupby(['Brand', 'Price Band']).agg({
    'Monthly Sales': 'mean',
    'Rating': 'mean',
    'Title': 'count'
}).rename(columns={'Title': 'Product Count'}).round(1)

print("\n📊 FIXED PRICE BAND SUMMARY:\n")
print(price_band_summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Function to estimate sales volume from strings like '6K+ bought in past month'
def estimate_sales_volume(sales_string):
    sales_string = str(sales_string).replace('bought in past month', '').strip()
    sales_string = sales_string.replace(',', '')
    if '+' in sales_string:
        sales_string = sales_string.replace('+', '')
        multiplier = 1
        if 'K' in sales_string:
            multiplier = 1000
            sales_string = sales_string.replace('K', '')
        sales = float(sales_string) * multiplier
    elif 'K' in sales_string:
        sales = float(sales_string.replace('K', '')) * 1000
    else:
        sales = float(sales_string)
    return sales

# Function to calculate average discount percentage
def calculate_average_discount(df):
    df['Price (INR)'] = pd.to_numeric(df['Price (INR)'], errors='coerce')
    df['Price Strikethrough (INR)'] = pd.to_numeric(df['Price Strikethrough (INR)'], errors='coerce')
    df = df.dropna(subset=['Price (INR)', 'Price Strikethrough (INR)'])
    df = df[df['Price Strikethrough (INR)'] != 0]
    df['Discount (%)'] = ((df['Price Strikethrough (INR)'] - df['Price (INR)']) / df['Price Strikethrough (INR)']) * 100
    return df['Discount (%)'].mean()

# Function to analyze trimmer performance
def analyze_trimmer_performance(df, brand_name):
    trimmer_df = df[df['Title'].str.contains('Trimmer', case=False)].copy()
    trimmer_df['Sales Volume'] = trimmer_df['Sales Volume'].astype(str)
    trimmer_df['Estimated Sales'] = trimmer_df['Sales Volume'].apply(estimate_sales_volume)
    trimmer_df['Price (INR)'] = pd.to_numeric(trimmer_df['Price (INR)'], errors='coerce')
    trimmer_df['Rating'] = pd.to_numeric(trimmer_df['Rating'], errors='coerce')
    average_price = trimmer_df['Price (INR)'].mean()
    average_discount = calculate_average_discount(trimmer_df)
    average_rating = trimmer_df['Rating'].mean()
    average_monthly_sales = trimmer_df['Estimated Sales'].mean()
    return average_price, average_discount, average_rating, average_monthly_sales

# Function to display competitive analysis as DataFrame
def display_competitive_analysis_df(c1, p1, d1, r1, s1, c2, p2, d2, r2, s2):
    data = {
        "Metric": ["Avg Price (INR)", "Avg Discount (%)", "Avg Rating", "Monthly Sales (avg)"],
        c1: [f"₹{p1:.1f}", f"{d1:.1f}%", f"⭐ {r1:.1f}", f"{s1:.0f} units"],
        c2: [f"₹{p2:.1f}", f"{d2:.1f}%", f"⭐ {r2:.1f}", f"{s2:.0f} units"],
        "Winner 🏆": [
            c1 if p1 < p2 else c2,
            c1 if d1 > d2 else c2,
            c1 if r1 > r2 else c2,
            c1 if s1 > s2 else c2,
        ]
    }
    df = pd.DataFrame(data)
    return df

# Function to plot bar chart for competitive analysis
def plot_comparative_bar_chart(c1, p1, d1, r1, s1, c2, p2, d2, r2, s2):
    metrics = ["Avg Price (INR)", "Avg Discount (%)", "Avg Rating", "Monthly Sales (avg)"]
    company1_values = [p1, d1, r1, s1]
    company2_values = [p2, d2, r2, s2]

    bar_width = 0.35
    index = range(len(metrics))

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(index, company1_values, bar_width, label=f"{c1}", color='skyblue')
    ax.bar([i + bar_width for i in index], company2_values, bar_width, label=f"{c2}", color='lightcoral')

    ax.set_xlabel('Metrics')
    ax.set_ylabel('Values')
    ax.set_title(f'Competitive Analysis: {c1} vs {c2}')
    ax.set_xticks([i + bar_width / 2 for i in index])
    ax.set_xticklabels(metrics)
    ax.legend()

    plt.tight_layout()
    plt.show()

# Sample CSV data
csv_data_company1 = """Title,Price (INR),Price Strikethrough (INR),Rating,Sales Volume
Trimmer A,999,1499,4.2,6K+ bought in past month
Trimmer B,899,1299,4.0,4K+ bought in past month
"""

csv_data_company2 = """Title,Price (INR),Price Strikethrough (INR),Rating,Sales Volume
Trimmer X,1099,1599,3.8,3K+ bought in past month
Trimmer Y,1199,1699,3.5,2K+ bought in past month
"""

company1_name = "Bombay Shaving Company"
company2_name = "Beardo"

# Load and analyze both CSVs
df1 = pd.read_csv(StringIO(csv_data_company1))
df2 = pd.read_csv(StringIO(csv_data_company2))

# Get analysis for both companies
p1, d1, r1, s1 = analyze_trimmer_performance(df1, company1_name)
p2, d2, r2, s2 = analyze_trimmer_performance(df2, company2_name)

# Display DataFrame for analysis
result_df = display_competitive_analysis_df(company1_name, p1, d1, r1, s1, company2_name, p2, d2, r2, s2)
print(result_df)

# Plot the comparative bar chart
plot_comparative_bar_chart(company1_name, p1, d1, r1, s1, company2_name, p2, d2, r2, s2)

# Strategic Insight
print(f"\n💡 Strategic Insight 💡\n")
print(f"In the trimmer category, {company1_name} outperforms {company2_name} across all key metrics on Amazon.")
print(f"{company1_name} trimmers are ~{abs(p2-p1)/p2*100:.1f}% more affordable and receive significantly better customer ratings (⭐ {r1:.1f} vs ⭐ {r2:.1f}).")
print(f"They also achieve {(s1 - s2)/s2 * 100:.1f}% higher average monthly sales.")
print(f"✅ This positions {company1_name} as a consumer-trusted, value-for-money leader in trimmers.")
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load your dataset
df = pd.read_csv("Reviews bsc.csv")

# Drop rows with missing reviews
df = df.dropna(subset=["Content"])

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Define sentiment classification function
def get_sentiment(text):
    score = analyzer.polarity_scores(str(text))["compound"]
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply to the Content column
df["VADER_Sentiment"] = df["Content"].apply(get_sentiment)

# See sentiment distribution
print(df["VADER_Sentiment"].value_counts())

# Optional: Save the result
df.to_csv("BSC_reviews_with_sentiment.csv", index=False)
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("BSC_reviews_with_sentiment.csv")

# Drop missing reviews
df = df.dropna(subset=["Content"])

# (Assuming you already have VADER_Sentiment column. If not, run sentiment classification first.)

# Function to generate and show word cloud
def generate_wordcloud(text, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(title, fontsize=16)
    plt.show()

# Create word clouds for each sentiment
for sentiment in ["Positive", "Negative", "Neutral"]:
    reviews = df[df["VADER_Sentiment"] == sentiment]["Content"]
    combined_text = " ".join(reviews.astype(str))
    generate_wordcloud(combined_text, f"{sentiment} Reviews Word Cloud")
```
